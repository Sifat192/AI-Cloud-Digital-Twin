```python
import pandas as pd
```

```python
df = pd.read_csv('cloud_dataset.csv')
```

```python
df.Timestamp = pd.to_datetime(df['Timestamp'])
df['Timestamp'].dtype
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Workload_Type'] = le.fit_transform(df['Workload_Type'])
df['User_ID'] = le.fit_transform(df['User_ID'])
df['Timestamp'] = df['Timestamp'].astype('int64') / 1e9
```

```python
x = df.drop('Anomaly_Label',axis=1)
y=df.Anomaly_Label
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape , x_test.shape , y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x_train = st.fit_transform(x_train)
x_test = st.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix,recall_score,precision_score
print("Accuracy:", accuracy_score(y_test,pred))
print("Precision:", precision_score(y_test,pred))
print("Recall:", recall_score(y_test,pred))
print("F1:", f1_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000,max_depth=20)
rf.fit(x_train,y_train)
pred = rf.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test,pred))
print("Precision:", precision_score(y_test,pred))
print("Recall:", recall_score(y_test,pred))
print("F1:", f1_score(y_test,pred))
```

```python
import joblib
joblib.dump(rf, "anomaly_model.pkl")
joblib.dump(st, "anomaly_scaler.pkl")
```
