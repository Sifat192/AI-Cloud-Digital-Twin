```python
import pandas as pd
```

```python
df = pd.read_csv('cloud_dataset.csv')
```

```python
df.Timestamp = pd.to_datetime(df['Timestamp'])
df['Timestamp'].dtype
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Workload_Type'] = le.fit_transform(df['Workload_Type'])
df['User_ID'] = le.fit_transform(df['User_ID'])
df['Timestamp'] = df['Timestamp'].astype('int64') / 1e9
```

```python
x = df.drop('CPU_Usage',axis=1)
y=df.CPU_Usage
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape , x_test.shape , y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x_train = st.fit_transform(x_train)
x_test = st.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print("MAE:", mean_absolute_error(y_test,pred))
print("MSE:", mean_squared_error(y_test,pred))
print("RMSE:", mean_squared_error(y_test,pred)**0.5)
print("R2:", r2_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=1000,max_depth=20)
rf.fit(x_train,y_train)
rf_pred = rf.predict(x_test)
```

```python
print("MAE:", mean_absolute_error(y_test,rf_pred))
print("MSE:", mean_squared_error(y_test,rf_pred))
print("RMSE:", mean_squared_error(y_test,rf_pred)**0.5)
print("R2:", r2_score(y_test,rf_pred))
```

```python
import joblib
joblib.dump(rf, "cpu_forecasting_model.pkl")
joblib.dump(st, "cpu_scaler.pkl")
```
